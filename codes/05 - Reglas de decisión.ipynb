# Reglas de decisión

## Ejemplo I

En este ejemplo, usamos el algoritmo Ripper para determinar si podemos sali a jugar, usando el popular dataset `weather`.

```python
# RIPPER no está implementado en Scikit-learn
# Instalamos wittgenstein (sólo si no lo tenemos instalado)
!pip install wittgenstein --upgrade
```

```python
# Importamos los módulos necesarios
import numpy as np
import pandas as pd
import wittgenstein as lw
from sklearn.metrics import confusion_matrix

# Leemos los datos que usaremos
weather =  pd.read_csv("https://raw.githubusercontent.com/jcobayliss/TC3002B-M2/refs/heads/main/datasets/weather.csv")

# Separamos las variables independientes de la dependiente
X = weather.drop("Play", axis = 1)
y = weather["Play"]

# Creamos un modelo de clasificación usando RIPPER
model = lw.RIPPER(random_state = 42)

# Entrenamos el modelo usando todos los datos disponibles (no recomendado, se hace con fines académicos)
model.fit(X, y, pos_class = "yes")

# Realizamos la predicción usando todos los datos disponibles (no recomendado, se hace con fines académicos)
prediction = model.predict(X)
prediction = np.where(prediction, "yes", "no")

# Generamos la matriz de confusión
confusion_matrix(y, prediction)

# Visualizamos el modelo (conjunto de reglas)
print(model.ruleset_)
```
