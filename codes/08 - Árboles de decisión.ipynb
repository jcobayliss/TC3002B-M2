# Árboles de decisión

```python
# Importamos los módulos necesarios
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler

# Generamos una función para visualizar los patrones en dos dimensiones
def plot2D(x1, x2, y, labels, caption):
  fig, ax = plt.subplots()
  for g in np.unique(y):
    i = np.where(y == g)[0]
    ax.scatter(x1[i], x2[i], color = labels[g], label = g)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title(caption)
  ax.legend()
```

## Ejemplo I

En este ejemplo, usamos un árbol de decisión para clasificar flores, usando el dataset clásico de iris.

```python
# Leemos los datos que usaremos
iris =  pd.read_csv("https://raw.githubusercontent.com/jcobayliss/TC3002B-M2/refs/heads/main/datasets/iris.csv")

# Separamos las variables independientes de la dependiente
X = iris.drop("Species", axis = 1)
y = iris["Species"]

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.60, stratify = y, random_state = 42)

# Creamos un modelo de clasificación usando un árbol de decisión
model = DecisionTreeClassifier(random_state = 42)

# Entrenamos el modelo usando los datos de entrenamiento
model.fit(X_train, y_train)

# Realizamos la predicción usando los datos de prueba
prediction = model.predict(X_test)

# Generamos la matriz de confusión
confusion_matrix(y_test, prediction)
```

```python
# Visualizamos el árbol de decisión generado
plt.figure(figsize=(10, 8))
plot_tree(
    model,
    feature_names = X.columns,
    class_names=[str(c) for c in model.classes_],
    filled = True,
    rounded = True,
    impurity = True,
    proportion = False,
    fontsize = 8
)
```

# Ejemplo II

Generalmente, los árboles de decisión son útiles no sólo por la clasificación en sí, sino porque permiten explicar cómo toman las decisiones. Sin embargo, hay que tener cuidado, porque cambios en el conjunto de entrenamiento pueden cambiar drásticamente la forma del árbol generado.

```python
# Leemos los datos que usaremos
iris =  pd.read_csv("https://raw.githubusercontent.com/jcobayliss/TC3002B-M2/refs/heads/main/datasets/iris.csv")

# Separamos las variables independientes de la dependiente
X = iris.drop("Species", axis = 1)
y = iris["Species"]

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.60, stratify = y, random_state = 24)

# Creamos un modelo de clasificación usando un árbol de decisión
model = DecisionTreeClassifier(random_state = 42)

# Entrenamos el modelo usando los datos de entrenamiento
model.fit(X_train, y_train)

# Realizamos la predicción usando los datos de prueba
prediction = model.predict(X_test)

# Generamos la matriz de confusión
confusion_matrix(y_test, prediction)
```

```python
# Visualizamos el árbol de decisión generado
plt.figure(figsize=(10, 8))
plot_tree(
    model,
    feature_names = X.columns,
    class_names=[str(c) for c in model.classes_],
    filled = True,
    rounded = True,
    impurity = True,
    proportion = False,
    fontsize = 8
)
```

## Ejemplo III

Una ventaja de los árboles de decisión es que no requieren que los datos cuantitativos se encuentren en rangos similares (los datos no requieren escalamiento). Este comportamiento puede observarse claramente en este ejemplo.

```python
# Leemos los datos que usaremos
iris =  pd.read_csv("https://raw.githubusercontent.com/jcobayliss/TC3002B-M2/refs/heads/main/datasets/vertical.csv")

# Separamos las variables independientes de la dependiente
X = iris.drop("Class", axis = 1)
y = iris["Class"]

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.60, stratify = y, random_state = 54321)

# Creamos un modelo de clasificación usando un árbol de decisión
model = DecisionTreeClassifier(random_state = 42)

# Entrenamos el modelo usando los datos de entrenamiento
model.fit(X_train, y_train)

# Realizamos la predicción usando los datos de prueba
prediction = model.predict(X_test)

# Generamos la matriz de confusión
confusion_matrix(y_test, prediction)
```

```python
# Asociamos un color a cada etiqueta de la clase
labels = {"up": "blue", "down": "red"}

# Visualizamos el patrón 2D de la realidad en el conjunto de entrenamiento
x1 = X_train.iloc[:, 0].values
x2 = X_train.iloc[:, 1].values
y = y_train
plot2D(x1, x2, y, labels, "Realidad (conjunto de entrenamiento)")

# Visualizamos el patrón 2D de la realidad en el conjunto de prueba
x1 = X_test.iloc[:, 0].values
x2 = X_test.iloc[:, 1].values
y = y_test
plot2D(x1, x2, y, labels, "Realidad (conjunto de prueba)")

# Visualizamos el patrón 2D de la predicción en el conjunto de prueba
x1 = X_test.iloc[:, 0].values
x2 = X_test.iloc[:, 1].values
y = prediction
plot2D(x1, x2, y, labels, "Predicción (conjunto de prueba)")
```

```python
# Visualizamos el árbol de decisión generado
plt.figure(figsize=(4, 4))
plot_tree(
    model,
    feature_names = X.columns,
    class_names=[str(c) for c in model.classes_],
    filled = True,
    rounded = True,
    impurity = True,
    proportion = False,
    fontsize = 8
)
```
